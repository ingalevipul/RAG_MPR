```python
import os
from dotenv import dotenv_values
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
import requests
import pandas as pd
from uuid import uuid4
from langchain.vectorstores import FAISS
from langchain.schema import Document
from dotenv import load_dotenv

config = dotenv_values(".env")
if not os.environ.get("GOOGLE_API_KEY"):
  os.environ["GOOGLE_API_KEY"] = config.get('gemini_api')


from langchain.chat_models import init_chat_model

model = init_chat_model("gemini-2.5-flash", model_provider="google_genai")


```

```python
from langchain_community.document_loaders import NewsURLLoader
def get_news(query="india tariff"):
    # Replace YOUR_NEWS_API_KEY as needed
    key=config.get('NEWS_API_KEY')
    
    # Expand search categories for comprehensive coverage
    categories = [
        "india+supply+chain",
        "india+natural+disasters",
        "india+geopolitics",
        "india+trade+policy",
        "india+infrastructure",
        "india+logistics",
        "india+manufacturing"
    ]
    
    all_articles = []
    for category in categories:
        url = f"https://newsapi.org/v2/everything?q={category}&sortBy=relevancy&apiKey={key}"
        resp = requests.get(url).json()
        articles = resp.get("articles", [])
        all_articles.extend([a['url'] for a in articles])
    
    # Remove duplicates while preserving order
    return list(dict.fromkeys(all_articles))

LL=get_news()
print(f"Total news articles found: {len(LL)}")

loader = NewsURLLoader(
    urls=LL,
)
docs = loader.load()
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200,length_function=len)


splits=text_splitter.split_documents(docs)
```

```python
Documents=[Document(page_content=i.page_content) for i in splits]
```

```python
from langchain_community.vectorstores import Chroma
import time

# Use existing ChromaDB
vectorstore = Chroma(
    persist_directory="./chromadb",
    embedding_function=embeddings
)

# Add documents in smaller batches
batch_size = 5
for i in range(0, len(Documents), batch_size):
    batch = Documents[i:i + batch_size]
    print(f"Adding batch {i//batch_size + 1} ({len(batch)} documents)")
    vectorstore.add_documents(batch)
    vectorstore.persist()  # Save after each batch
    if i + batch_size < len(Documents):
        time.sleep(5)  # Add a small delay between batches to avoid rate limits

print(f"Finished adding {len(Documents)} documents in {(len(Documents)-1)//batch_size + 1} batches")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.schema.runnable import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

template = """You are an AI system for Proactive Supply Chain Risk Assessment.
You will be given retrieved context from news articles related to India, context:{context}
Question:{question}

Your responsibilities are:
1. Ground your analysis in the retrieved articles. Do not make up facts.

2. Risk Identification and Classification:
   - Natural disasters (earthquakes, floods, extreme weather)
   - Geopolitical events (trade tensions, conflicts, policy changes)
   - Infrastructure issues (port congestion, transportation)
   - Market dynamics (demand fluctuations, price volatility)
   - Operational risks (labor issues, technology disruptions)
   - Compliance & regulatory risks
   - Environmental & sustainability risks

3. Risk Analysis:
   - Likelihood of occurrence (High/Medium/Low)
   - Potential timing (Immediate/Short-term/Long-term)
   - Geographic scope (Local/Regional/Global)
   - Impact severity (Critical/Moderate/Minor)

4. Impact Assessment Dimensions:
   - Supply chain disruptions (delays, bottlenecks)
   - Financial impact (costs, revenue loss)
   - Operational impact (production, logistics)
   - Market impact (demand, competition)
   - Stakeholder impact (suppliers, customers)

5. Mitigation Strategy Framework:
   - Immediate response actions
   - Preventive measures
   - Alternative sourcing options
   - Buffer strategies (inventory, capacity)
   - Technology solutions
   - Relationship management
   - Monitoring & early warning systems

Output Format:
Risk Category:
Risk Level: (High/Medium/Low)
Timeframe: (Immediate/Short-term/Long-term)
Geographic Scope: (Local/Regional/Global)

Retrieved Evidence:
[Key facts from articles with specific details]

Risk Scenario Analysis:
[Detailed scenario based on evidence]

Multi-dimensional Impact Assessment:
- Supply Chain Impact:
- Financial Impact:
- Operational Impact:
- Market Impact:
- Stakeholder Impact:

Mitigation Strategy:
1. Immediate Actions:
2. Preventive Measures:
3. Alternative Solutions:
4. Monitoring Plan:

Rules:
- Ground all analysis in provided evidence
- Quantify impacts where possible
- Provide specific, actionable recommendations
- Highlight uncertainties and assumptions
- Consider interconnected effects
"""

prompt = ChatPromptTemplate.from_template(template)

def docs2str(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": retriever | docs2str, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timedelta
import numpy as np

class RiskAlertSystem:
    def __init__(self, vectorstore, threshold_scores={'High': 0.8, 'Medium': 0.5, 'Low': 0.3}):
        self.vectorstore = vectorstore
        self.threshold_scores = threshold_scores
        self.alerts = []
    
    def analyze_risk(self, query):
        # Get relevant documents
        results = self.vectorstore.similarity_search_with_score(query, k=5)
        
        # Calculate risk score based on similarity scores
        avg_score = np.mean([score for _, score in results])
        
        # Determine risk level
        risk_level = 'Low'
        for level, threshold in self.threshold_scores.items():
            if avg_score >= threshold:
                risk_level = level
                break
        
        # Create alert if risk is Medium or High
        if risk_level in ['Medium', 'High']:
            alert = {
                'timestamp': datetime.now(),
                'query': query,
                'risk_level': risk_level,
                'score': avg_score,
                'evidence': [doc.page_content for doc, _ in results]
            }
            self.alerts.append(alert)
        
        return risk_level, avg_score, results

    def get_recent_alerts(self, hours=24):
        cutoff = datetime.now() - timedelta(hours=hours)
        return [alert for alert in self.alerts if alert['timestamp'] > cutoff]

# Initialize the alert system
alert_system = RiskAlertSystem(vectorstore)

# Test the system with different queries
risk_queries = [
    "natural disasters in India",
    "trade policy changes India",
    "infrastructure problems India ports",
    "supply chain disruptions India",
    "geopolitical tensions India"
]

results_data = []
for query in risk_queries:
    risk_level, risk_score, _ = alert_system.analyze_risk(query)
    results_data.append({
        'query': query,
        'risk_level': risk_level,
        'risk_score': risk_score
    })
```

```python
# Create interactive dashboard
def create_risk_dashboard(results_data, alerts):
    # 1. Risk Level Distribution
    risk_counts = pd.DataFrame(results_data).groupby('risk_level').size().reset_index(name='count')
    fig1 = px.pie(risk_counts, values='count', names='risk_level', 
                  title='Distribution of Risk Levels',
                  color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
    
    # 2. Risk Scores by Category
    fig2 = px.bar(results_data, x='query', y='risk_score',
                  title='Risk Scores by Category',
                  color='risk_level',
                  color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
    fig2.update_layout(xaxis_title='Category', yaxis_title='Risk Score')
    
    # 3. Alert Timeline
    if alerts:
        alert_df = pd.DataFrame(alerts)
        fig3 = px.scatter(alert_df, x='timestamp', y='score',
                         color='risk_level',
                         title='Alert Timeline',
                         color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
        fig3.update_layout(xaxis_title='Time', yaxis_title='Risk Score')
    
    # Display all figures
    fig1.show()
    fig2.show()
    if alerts:
        fig3.show()

# Get recent alerts
recent_alerts = alert_system.get_recent_alerts()

# Create and display the dashboard
create_risk_dashboard(results_data, recent_alerts)

# Create a summary table of current risks
risk_summary = pd.DataFrame(results_data)
print("\nRisk Summary Table:")
print(risk_summary.to_string(index=False))
```

# Supply Chain Risk Monitoring Dashboard

This dashboard provides real-time monitoring of supply chain risks with the following components:

1. **Risk Level Distribution**: Pie chart showing the distribution of risk levels across different categories
2. **Risk Scores by Category**: Bar chart displaying risk scores for each monitored category
3. **Alert Timeline**: Scatter plot showing alerts over time with their risk levels
4. **Risk Summary Table**: Detailed view of current risk assessments

The system automatically:
- Monitors multiple risk categories
- Generates alerts for medium and high-risk events
- Provides visual analytics for risk assessment
- Maintains a timeline of identified risks

```python
import gradio as gr
import plotly.io as pio

class SupplyChainRiskDashboard:
    def __init__(self, alert_system, rag_chain):
        self.alert_system = alert_system
        self.rag_chain = rag_chain
        
    def analyze_new_risk(self, query):
        # Get risk analysis
        risk_level, risk_score, results = self.alert_system.analyze_risk(query)
        
        # Get RAG response
        rag_response = self.rag_chain.invoke(query)
        
        # Create visualizations
        results_data = [{'query': query, 'risk_level': risk_level, 'risk_score': risk_score}]
        
        # Get recent alerts
        recent_alerts = self.alert_system.get_recent_alerts()
        
        # Generate plots
        fig1, fig2, fig3 = self.create_dashboard_plots(results_data, recent_alerts)
        
        # Parse RAG response into sections
        risk_summary = f"""
        🎯 Query: {query}
        🔔 Risk Level: {risk_level}
        📊 Risk Score: {risk_score:.2f}
        """
        
        # Extract sections from RAG response
        sections = rag_response.split('\n\n')
        evidence = next((s for s in sections if 'Retrieved Evidence:' in s), '')
        impact = next((s for s in sections if 'Multi-dimensional Impact Assessment:' in s), '')
        mitigation = next((s for s in sections if 'Mitigation Strategy:' in s), '')
        
        return (fig1, fig2, fig3 if fig3 else None,
                risk_summary, evidence, impact, mitigation)
    
    def create_dashboard_plots(self, results_data, alerts):
        # Risk Level Distribution
        risk_counts = pd.DataFrame(results_data).groupby('risk_level').size().reset_index(name='count')
        fig1 = px.pie(risk_counts, values='count', names='risk_level',
                      title='Risk Level Distribution',
                      color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
        fig1.update_layout(height=400)
        
        # Risk Scores
        fig2 = px.bar(results_data, x='query', y='risk_score',
                      title='Risk Score Analysis',
                      color='risk_level',
                      color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
        fig2.update_layout(height=400)
        
        # Alert Timeline
        fig3 = None
        if alerts:
            alert_df = pd.DataFrame(alerts)
            fig3 = px.scatter(alert_df, x='timestamp', y='score',
                            color='risk_level',
                            title='Alert Timeline',
                            color_discrete_map={'High': 'red', 'Medium': 'yellow', 'Low': 'green'})
            fig3.update_layout(height=400)
        
        return fig1, fig2, fig3

# Create dashboard instance
dashboard = SupplyChainRiskDashboard(alert_system, rag_chain)

# Create the Gradio app with Blocks
with gr.Blocks(theme=gr.themes.Soft()) as app:
    gr.Markdown("# 🏭 Supply Chain Risk Assessment Dashboard")
    
    with gr.Row():
        query_input = gr.Textbox(
            label="Enter your supply chain risk query",
            placeholder="E.g., What are the current infrastructure risks in Indian ports?",
            lines=2
        )
    
    with gr.Row():
        analyze_btn = gr.Button("Analyze Risk", variant="primary")
    
    with gr.Tabs():
        with gr.TabItem("📊 Risk Visualization"):
            with gr.Row(equal_height=True):
                with gr.Column():
                    risk_dist_plot = gr.Plot(label="Risk Level Distribution")
                with gr.Column():
                    risk_score_plot = gr.Plot(label="Risk Score Analysis")
            with gr.Row():
                alert_timeline = gr.Plot(label="Alert Timeline")
        
        with gr.TabItem("📝 Analysis Details"):
            with gr.Row():
                with gr.Column():
                    risk_summary = gr.Textbox(label="Risk Summary", lines=4)
                with gr.Column():
                    evidence = gr.Markdown(label="Retrieved Evidence")
            with gr.Row():
                with gr.Column():
                    impact = gr.Markdown(label="Impact Assessment")
                with gr.Column():
                    mitigation = gr.Markdown(label="Mitigation Strategy")

    analyze_btn.click(
        fn=dashboard.analyze_new_risk,
        inputs=[query_input],
        outputs=[risk_dist_plot, risk_score_plot, alert_timeline,
                risk_summary, evidence, impact, mitigation]
    )

    gr.Markdown("""
    ### 📋 Instructions
    1. Enter your query about potential supply chain risks in India
    2. Click 'Analyze Risk' to get comprehensive analysis
    3. View visualizations in the 'Risk Visualization' tab
    4. Check detailed analysis in the 'Analysis Details' tab
    
    ### 🔍 Suggested Queries
    - What are the current infrastructure risks in Indian ports?
    - Analyze potential geopolitical risks affecting Indian supply chains
    - Assess natural disaster risks in major Indian manufacturing hubs
    """)

# Launch the interface
app.launch(share=True)
```
